```python
import os, copy, time, random, itertools
from pathlib import Path
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from tqdm.notebook import tqdm

DATA_ROOT = "/content/drive/MyDrive/Project Data/Fruit"
TRAIN_DIR = os.path.join(DATA_ROOT, "Train")
VAL_DIR   = os.path.join(DATA_ROOT, "Validation")
BATCH_SIZE = 32
IMAGE_SIZE = 224
NUM_EPOCHS = 15
LR = 1e-4
WEIGHT_DECAY = 1e-4
NUM_WORKERS = 2
SAVE_PATH = "/content/drive/MyDrive/best_fruit_resnet50.pth"
CONF_MATRIX_PATH = "/content/drive/MyDrive/confusion_matrix.png"
VISUAL_PATH = "/content/drive/MyDrive/validation_visualization.png"
USE_PRETRAINED = True
USE_AMP = True
SEED = 42

random.seed(SEED)
torch.manual_seed(SEED)
np.random.seed(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)
if torch.cuda.is_available():
    try:
        print("GPU:", torch.cuda.get_device_name(0))
    except Exception:
        pass

torch.backends.cudnn.benchmark = True

train_transforms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(12),
    transforms.ColorJitter(0.2, 0.2, 0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])
val_transforms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])

IMG_EXTS = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')
def is_image_file(filename):
    return filename.lower().endswith(IMG_EXTS)

class ImagesSubfolderDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_to_idx = {cls_name: idx for idx, cls_name in enumerate(self.classes)}
        self.samples = []
        for cls in self.classes:
            cls_path = os.path.join(root_dir, cls)
            images_dir = os.path.join(cls_path, "Images")
            search_path = images_dir if os.path.isdir(images_dir) else cls_path
            for fname in sorted(os.listdir(search_path)):
                full = os.path.join(search_path, fname)
                if os.path.isfile(full) and is_image_file(full):
                    self.samples.append((full, self.class_to_idx[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, label

train_dataset = ImagesSubfolderDataset(TRAIN_DIR, transform=train_transforms)
val_dataset   = ImagesSubfolderDataset(VAL_DIR, transform=val_transforms)

if len(train_dataset) == 0 or len(val_dataset) == 0:
    raise RuntimeError("Empty dataset. Check directories.")

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                        num_workers=NUM_WORKERS, pin_memory=True)

num_classes = len(train_dataset.classes)
print(f"Found {num_classes} classes: {train_dataset.classes}")
print(f"Train samples: {len(train_dataset)}, Val samples: {len(val_dataset)}")

model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1 if USE_PRETRAINED else None)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)
scaler = torch.cuda.amp.GradScaler(enabled=USE_AMP and torch.cuda.is_available())

def train_one_epoch(model, loader, criterion, optimizer, device, scaler=None):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    n = 0
    pbar = tqdm(loader, desc="Train batches", leave=False)
    for inputs, labels in pbar:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        if scaler is not None:
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        preds = outputs.argmax(dim=1)
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels).item()
        n += inputs.size(0)
        pbar.set_postfix(loss=running_loss/n, acc=100*running_corrects/n)
    return running_loss/n, running_corrects/n

def eval_one_epoch(model, loader, criterion, device, return_preds_and_labels=False):
    model.eval()
    running_loss = 0.0
    running_corrects = 0
    n = 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        pbar = tqdm(loader, desc="Val batches", leave=False)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels).item()
            n += inputs.size(0)
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
            pbar.set_postfix(val_loss=running_loss/n, val_acc=100*running_corrects/n)
    if return_preds_and_labels:
        return running_loss/n, running_corrects/n, np.concatenate(all_preds), np.concatenate(all_labels)
    return running_loss/n, running_corrects/n

def plot_confusion_matrix(cm, class_names, normalize=True, cmap=plt.cm.Blues, out_path=None):
    if normalize:
        cm_sum = cm.sum(axis=1)[:, np.newaxis]
        cm = np.divide(cm, cm_sum, where=cm_sum!=0)
    plt.figure(figsize=(12,10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title("Confusion Matrix" + (" (normalized)" if normalize else ""))
    plt.colorbar(fraction=0.046, pad=0.04)
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=90, fontsize=8)
    plt.yticks(tick_marks, class_names, fontsize=8)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max()/2 if cm.max()!=0 else 0.5
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        val = cm[i,j]
        display = format(val, '.2f') if normalize else format(int(val),'d')
        plt.text(j,i,display,horizontalalignment="center",
                 color="white" if val>thresh else "black", fontsize=6)
    plt.tight_layout()
    if out_path:
        plt.savefig(out_path, dpi=200)
        print("Saved confusion matrix to:", out_path)
    plt.close()

best_acc = 0.0
best_model_wts = copy.deepcopy(model.state_dict())
total_start_time = time.time()

for epoch in range(1, NUM_EPOCHS+1):
    print(f"\nEpoch {epoch}/{NUM_EPOCHS}")
    start_time = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE, scaler if USE_AMP else None)
    train_time = time.time() - start_time

    start_val = time.time()
    val_loss, val_acc = eval_one_epoch(model, val_loader, criterion, DEVICE)
    val_time = time.time() - start_val

    print(f" Train Loss: {train_loss:.4f}  Acc: {100*train_acc:.2f}%  Time: {train_time:.2f}s")
    print(f" Val   Loss: {val_loss:.4f}  Acc: {100*val_acc:.2f}%  Time: {val_time:.2f}s")

    if val_acc > best_acc:
        best_acc = val_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save({
            'model_state_dict': best_model_wts,
            'best_acc': best_acc,
            'classes': train_dataset.classes
        }, SAVE_PATH)
        print(f" Saved best model -> {SAVE_PATH}")

total_time = time.time() - total_start_time
print(f"\nTotal training time: {total_time:.2f}s")

_, _, preds_all, labels_all = eval_one_epoch(model, val_loader, criterion, DEVICE, return_preds_and_labels=True)
cm = confusion_matrix(labels_all, preds_all)
plot_confusion_matrix(cm, train_dataset.classes, normalize=True, out_path=CONF_MATRIX_PATH)
print(f"Saved confusion matrix -> {CONF_MATRIX_PATH}")

images_to_show = 16
plt.figure(figsize=(12,12))
shown = 0
with torch.no_grad():
    for imgs, labels in val_loader:
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        preds = model(imgs).argmax(1)
        for i in range(len(imgs)):
            if shown >= images_to_show:
                break
            img = imgs[i].cpu().permute(1,2,0)
            gt = train_dataset.classes[labels[i].item()]
            pr = train_dataset.classes[preds[i].item()]
            color = "green" if gt==pr else "red"
            plt.subplot(4,4,shown+1)
            plt.imshow(img)
            plt.title(f"GT: {gt}\nPred: {pr}", color=color, fontsize=8)
            plt.axis("off")
            shown += 1
        if shown >= images_to_show:
            break
plt.tight_layout()
plt.savefig(VISUAL_PATH, dpi=200)
plt.show()
print(f"Saved validation visualization -> {VISUAL_PATH}")
```
