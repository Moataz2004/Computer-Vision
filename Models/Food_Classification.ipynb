```python
import os
import random
from collections import defaultdict
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image

import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import itertools
import time


DATA_ROOT = "/content/drive/MyDrive/Project Data/Food"
TRAIN_FOLDER = os.path.join(DATA_ROOT, "Train")
VAL_FOLDER = os.path.join(DATA_ROOT, "Validation")
SAVE_DIR = "/content/drive/MyDrive/temp/results_triplet_amp"
os.makedirs(SAVE_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 16
EMBEDDING_SIZE = 256
LR = 1e-4
WEIGHT_DECAY = 1e-6
EPOCHS = 20
MARGIN = 0.3
NUM_WORKERS = 4
KNN_K = 5
SAVE_INTERVAL = 5


def pil_loader(path):
    return Image.open(path).convert("RGB")


def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues, out_file=None):
    if normalize:
        cm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-12)
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    if len(classes) <= 30:
        plt.xticks(tick_marks, classes, rotation=90)
        plt.yticks(tick_marks, classes)
    else:
        plt.xticks([])
        plt.yticks([])
    fmt = '.2f' if normalize else 'd'
    thresh = (cm.max() + cm.min()) / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if len(classes) <= 30:
            plt.text(j, i, format(cm[i, j], fmt),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    if out_file:
        plt.savefig(out_file, bbox_inches='tight')
    plt.show()


class TripletFolderDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_to_idx = {c: i for i, c in enumerate(self.classes)}
        self.samples = []
        self.class_samples = defaultdict(list)
        for c in self.classes:
            cpath = os.path.join(root_dir, c)
            files = sorted([os.path.join(cpath, f) for f in os.listdir(cpath) if f.lower().endswith((".jpg",".jpeg",".png"))])
            if len(files) == 0:
                continue
            idx = self.class_to_idx[c]
            self.class_samples[idx] = files
            for f in files:
                self.samples.append((f, idx))
        self.classes_with_two = [k for k,v in self.class_samples.items() if len(v) >= 2]
        if len(self.classes_with_two) == 0:
            raise ValueError("No class has >= 2 images - can't form positive pairs.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        anchor_path, anchor_label = self.samples[idx]
        pos_candidates = self.class_samples[anchor_label]
        pos_path = anchor_path
        while pos_path == anchor_path:
            pos_path = random.choice(pos_candidates)
        neg_label = anchor_label
        while neg_label == anchor_label:
            neg_label = random.choice(list(self.class_samples.keys()))
        neg_path = random.choice(self.class_samples[neg_label])
        anchor = pil_loader(anchor_path)
        positive = pil_loader(pos_path)
        negative = pil_loader(neg_path)
        if self.transform:
            anchor = self.transform(anchor)
            positive = self.transform(positive)
            negative = self.transform(negative)
        return (anchor, positive, negative), (anchor_label, anchor_label, neg_label)


class SimpleImageFolderDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_to_idx = {c:i for i,c in enumerate(classes)}
        for c in classes:
            cpath = os.path.join(root_dir, c)
            files = sorted([os.path.join(cpath, f) for f in os.listdir(cpath) if f.lower().endswith((".jpg",".jpeg",".png"))])
            for f in files:
                self.samples.append((f, self.class_to_idx[c]))
    def __len__(self):
        return len(self.samples)
    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = pil_loader(path)
        if self.transform:
            img = self.transform(img)
        return img, label


class EmbeddingNet(nn.Module):
    def __init__(self, embedding_size=256, pretrained=True):
        super().__init__()
        backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2 if pretrained else None)
        modules = list(backbone.children())[:-1]
        self.backbone = nn.Sequential(*modules)
        self.fc = nn.Linear(2048, embedding_size)

    def forward(self, x):
        x = self.backbone(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        x = F.normalize(x, p=2, dim=1)
        return x


def compute_embeddings(dataloader, model, device):
    model.eval()
    embs = []
    labels = []
    with torch.no_grad():
        for imgs, labs in tqdm(dataloader, desc="embed"):
            imgs = imgs.to(device)
            out = model(imgs)
            embs.append(out.cpu().numpy())
            labels.extend(labs)
    embs = np.vstack(embs)
    labels = np.array(labels)
    return embs, labels


def train_epoch_amp(model, loader, optimizer, criterion, device, scaler):
    model.train()
    running_loss = 0.0
    pbar = tqdm(loader, desc="train")
    for (anc, pos, neg), _labels in pbar:
        anc, pos, neg = anc.to(device), pos.to(device), neg.to(device)
        optimizer.zero_grad()
        with torch.cuda.amp.autocast():
            emb_a = model(anc)
            emb_p = model(pos)
            emb_n = model(neg)
            loss = criterion(emb_a, emb_p, emb_n)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        running_loss += loss.item() * anc.size(0)
        pbar.set_postfix(loss=loss.item())
    return running_loss / len(loader.dataset)


def save_checkpoint(model, optimizer, scaler, epoch, save_path, is_best=False):
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'scaler_state_dict': scaler.state_dict(),
    }
    torch.save(checkpoint, save_path)
    if is_best:
        print(f"Best model saved at epoch {epoch}: {save_path}")


def main():
    total_start_time = time.time()

    train_transform = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ColorJitter(0.2,0.2,0.2,0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])
    eval_transform = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])

    train_dataset = TripletFolderDataset(TRAIN_FOLDER, transform=train_transform)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, drop_last=True)
    val_dataset = SimpleImageFolderDataset(VAL_FOLDER, transform=eval_transform)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=NUM_WORKERS)

    print(f"Train classes: {len(train_dataset.class_to_idx)}, Val classes: {len(val_dataset.class_to_idx)}")

    model = EmbeddingNet(embedding_size=EMBEDDING_SIZE, pretrained=True).to(DEVICE)
    criterion = nn.TripletMarginLoss(margin=MARGIN, p=2)
    optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.5)
    scaler = torch.cuda.amp.GradScaler()

    best_val_acc = 0.0

    for epoch in range(1, EPOCHS+1):
        print(f"\nEpoch {epoch}/{EPOCHS}")

        train_start = time.time()
        train_loss = train_epoch_amp(model, train_loader, optimizer, criterion, DEVICE, scaler)
        train_end = time.time()
        print(f"Train loss: {train_loss:.4f}  TrainTime={train_end-train_start:.2f}s")
        scheduler.step()

        val_start = time.time()
        val_embs, val_labels = compute_embeddings(val_loader, model, DEVICE)
        n = len(val_labels)
        idxs = np.arange(n)
        np.random.shuffle(idxs)
        split = int(0.8 * n)
        gallery_idx = idxs[:split]
        query_idx = idxs[split:]
        knn = KNeighborsClassifier(n_neighbors=min(KNN_K, len(gallery_idx)))
        knn.fit(val_embs[gallery_idx], val_labels[gallery_idx])
        preds = knn.predict(val_embs[query_idx])
        acc = accuracy_score(val_labels[query_idx], preds)
        val_end = time.time()
        print(f"Validation KNN accuracy: {acc:.4f}  ValTime={val_end-val_start:.2f}s")

        best_model_path = os.path.join(SAVE_DIR, "best_embedding_model.pth")
        if acc > best_val_acc:
            best_val_acc = acc
            save_checkpoint(model, optimizer, scaler, epoch, best_model_path, is_best=True)

        if epoch % SAVE_INTERVAL == 0:
            ckpt_path = os.path.join(SAVE_DIR, f"checkpoint_epoch_{epoch}.pth")
            save_checkpoint(model, optimizer, scaler, epoch, ckpt_path)
            print(f"Saved checkpoint: {ckpt_path}")

    total_end_time = time.time()
    print(f"\nTotal training time: {total_end_time-total_start_time:.2f}s")
    print(f"Best validation KNN accuracy: {best_val_acc:.4f}")

    N = 5
    model.eval()
    val_embs, val_labels = compute_embeddings(val_loader, model, DEVICE)
    n = len(val_labels)
    idxs = np.arange(n)
    np.random.shuffle(idxs)
    split = int(0.8 * n)
    gallery_idx = idxs[:split]
    query_idx = idxs[split:]
    knn = KNeighborsClassifier(n_neighbors=min(KNN_K, len(gallery_idx)))
    knn.fit(val_embs[gallery_idx], val_labels[gallery_idx])
    preds = knn.predict(val_embs[query_idx])

    idx_to_class = {v:k for k,v in val_dataset.class_to_idx.items()}
    for i in range(min(N, len(query_idx))):
        img = val_dataset[query_idx[i]][0]
        true_name = idx_to_class[val_labels[query_idx[i]]]
        pred_name = idx_to_class[preds[i]]
        plt.figure(figsize=(6,3))
        plt.subplot(1,2,1)
        plt.imshow(img.permute(1,2,0).numpy() * 0.229 + 0.485)
        plt.title(f"True: {true_name}")
        plt.axis('off')
        plt.subplot(1,2,2)
        plt.imshow(img.permute(1,2,0).numpy() * 0.229 + 0.485)
        plt.title(f"Pred: {pred_name}")
        plt.axis('off')
        plt.tight_layout()
        plt.show()

    final_model_path = os.path.join(SAVE_DIR, "final_embedding_model.pth")
    save_checkpoint(model, optimizer, scaler, EPOCHS, final_model_path)
    print(f"Final model saved: {final_model_path}")


if __name__ == "__main__":
    main()
```
